```python
# to delete missing values in customer id column 

df=df.dropna(subset=['CustomerID'])
```

```python
# to remove negetive values
df=df[df['Quantity']>0]
```

```python
# to remove negetive price
df=df[df['UnitPrice']>0]
```

```python
# revenue column

df['Revenue']=df['Quantity']*df['UnitPrice']
```

```python
# date convert

df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
grp=df.groupby('CustomerID').agg({'Revenue':'sum',
                                 'InvoiceNo': 'nunique',
                                  'InvoiceDate': 'max'
                
                                 }).reset_index()
```

```python
grp.columns=['CustomerID','TotalRevenue','TotalTransactions','LastPurchaseDate']
```

```python
df.info()
```

```python
grp.head()
```

# RFM Feature Engineering

```python
reference_date=grp['LastPurchaseDate'].max()+pd.Timedelta(days=1)
```

```python
grp['Recency']=(reference_date - grp['LastPurchaseDate']).dt.days
```

```python
grp['frequency']=grp['TotalTransactions']
```

```python
grp['Monetary']=grp['TotalRevenue']
```

```python
rfm=grp[['CustomerID','Recency','frequency','Monetary']]
```

```python
rfm.head()
```

# EDA + Business Insights

# RFM Scoring

```python
rfm['F_score']=pd.qcut(rfm['frequency'],q=5,duplicates='drop')
```

```python
rfm.head()
```

# Monetary Scoring

```python
rfm['M_score']=pd.qcut(rfm['Monetary'],q=5,duplicates='drop')
```

```python
rfm.head()
```

```python
rfm[['Monetary','M_score']].head()
rfm['M_score'].value_counts()
```

# Recency score

```python
rfm['R_score']=pd.qcut(rfm['Recency'],q=5,labels=[5,4,3,2,1],duplicates='drop')
```

```python
rfm[['Recency','R_score']].head()
rfm['R_score'].value_counts()
```

# final RFM Score combine

```python
rfm['R_score']=rfm['R_score'].astype(str)
rfm['M_score']=rfm['M_score'].astype(str)
rfm['F_score']=rfm['F_score'].astype(str)

```

```python
rfm['RFM_Score']=rfm['R_score']+rfm['M_score']+rfm['F_score']
```

# Basic Segmentation

```python
def segment_customer(row):
    if row['RFM_Score']=='555':
        return 'VIP'
    elif row['F_score']>='4' and row['M_score']>='4':
        return 'Loyal'
    elif row['R_score']=='1':
        return 'At Risk'
    else:
        return 'Regular'
```

```python
rfm['Segment']=rfm.apply(segment_customer , axis=1)
```

```python
rfm.head()
```

# engagement Score

```python
# normalize recency

rfm['Recency_norm']=(rfm['Recency'].max()-rfm['Recency'])/(rfm['Recency'].max()-rfm['Recency'].min())
```

```python
# frequency normalize

rfm['frequency_norm']=(rfm['frequency'].max()-rfm['frequency'])/(rfm['frequency'].max()-rfm['frequency'].min())
```

```python
# Engagement score

rfm['EngagementScore']=(rfm['Recency_norm']+rfm['frequency_norm'])/2
```

```python
rfm[['Recency','frequency','EngagementScore']].head()
```

# Avg purchase gap

```python
# data Sort

df=df.sort_values(by=['CustomerID','InvoiceDate'])
```

```python
# previous purchase date

df['PrevDate']=df.groupby('CustomerID')['InvoiceDate'].shift(1)
```

```python
# gap calculate

df['Gap']=(df['InvoiceDate']-df['PrevDate']).dt.days
```

```python
# avg gap per customer

avg_gap=df.groupby('CustomerID')['Gap'].mean().reset_index()
```

```python
# Merge RFM

rfm=rfm.merge(avg_gap , on='CustomerID',how='left')
```

```python
rfm.head()
```

# Customer activity trends

```python
# cutoff date 

cutoff_date=df['InvoiceDate'].max()-pd.Timedelta(days=90)
```

```python
# recent vs old transaction account

recent=df[df['InvoiceDate']
>
cutoff_date].groupby('CustomerID').size().reset_index(name='RecentTx')


old=df[df['InvoiceDate']<=
cutoff_date].groupby('CustomerID').size().reset_index(name='OldTx')
```

```python
# Merge

trend=recent.merge(old,on='CustomerID' , how='outer').fillna(0)
```

```python
# trend

trend['ActivityTrend']=trend['RecentTx']-trend['OldTx']

```

```python
# merge Rfm

rfm=rfm.merge(trend[['CustomerID','ActivityTrend']],on='CustomerID',how='left')
```

```python
rfm.head()
```

# Churn Prediction

```python
rfm['Churn']=(rfm['Recency']>90).astype(int)
```
